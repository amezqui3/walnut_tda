```python
import pandas as pd
import numpy as np
```

```python
src = '../hpcc/traditional/'
filename = '../all_phenos_for_Erik_221024.csv'
attr = pd.read_csv(filename)
print(attr.shape)
```

```python
col_names = ['bname', 'wname',
             'nut_length','nut_height', 'nut_width',
             'nut_vol',
             'nut_va3d',
             'nut_ratio_feret',
             'nut_area',
             'nut_sphericity',
             'nut_cnvx_area',
             'nut_cnvx_vol',
             'nut_ratio_cnvx_area',
             'nut_inv_ratio_cnvx_vol',
             'krumbein',
             'sneed',
             'air_vol', 'meat_vol', 'shell_vol', 'vein_vol',
             'air_ratio_vol', 'meat_ratio_vol', 'shell_ratio_vol', 'vein_ratio_vol',
             'shell_rugosity',
             'shell_thickness',
             'shell_ext_ratio', 'shell_prot_ratio',
             'shell_ext_vol', 'shell_prot_vol',
             'meat_length', 'meat_height', 'meat_width',
             'meat_area',
             'meat_cnvx_vol',
             'meat_cnvx_area',
             'meat_ratio_cnvx_area',
             'meat_inv_ratio_cnvx_vol',
             'meat_cavity_area',
             'meat_cavity_vol',
             'meat_arc_height',
             'meat_cavity_height',
             'meat_sphericity',
             'meat_ratio_cavity_area',
             'meat_ratio_cavity_vol',
             'rho_mvs', 'rho_vvs', 'rho_vvm'
             ]

filename = src + 'traditional_raw.csv'
raw = pd.read_csv(filename, header=None, dtype={1:str}, names=col_names)
raw.loc[raw['bname'] == '201SBa_R5_T55', 'bname'] = '2014SBa_R5_T55'
foo = raw.iloc[:, 30:33].values
MM = np.max(foo, axis=1)
mm = np.min(foo, axis=1)
raw['meat_ratio_feret'] = MM/mm

raw['nut_vol_sa'] = raw['meat_vol'] + raw['vein_vol'] + raw['shell_vol']

raw['meat_ratio_vol_sa'] = raw['meat_vol']/raw['nut_vol_sa']
raw['vein_ratio_vol_sa'] = raw['vein_vol']/raw['nut_vol_sa']
raw['shell_ratio_vol_sa'] = raw['shell_vol']/raw['nut_vol_sa']

raw['meat_ratio_cnvx_area'] = 1/raw['meat_ratio_cnvx_area']
raw['meat_inv_ratio_cnvx_vol'] = 1/raw['meat_inv_ratio_cnvx_vol']
raw = raw.drop(columns=['shell_ext_ratio', 'shell_ext_vol'])

#data = data.drop(columns = [''])
print(raw.shape)
raw.head()
```

```python
cattraits = np.zeros((len(raw), 15)) - 1
UCACCSD = []

for i in range(len(cattraits)):
    locate, Rrow, Ttree = raw.iloc[i,0].split('_')
    row = int(Rrow[1:])
    tree = int(Ttree[1:])

    if locate == 'NewStuke':
        locate = 'New Stuke'
    elif locate in ['SelectD', 'SelectF', 'SelectC']:
        locate = 'Select Davis'

    metadf = attr[(attr['LOCATE']==locate) & (attr['ROW']==row) & (attr['TREE']==tree)]
    if len(metadf) > 0:
        ucaccsd = metadf['UCACCSD'].iloc[0]
    else:
        ucaccsd = ''
        print(raw.iloc[i,0])
    
    UCACCSD.append(ucaccsd)
```

```python
raw['UCACCSD'] = UCACCSD
foo = np.asarray([51] + list(range(51)))
raw = raw.iloc[:, foo]
raw.head()
```

```python
kratio = raw.loc[:,'meat_ratio_vol'].values
q25,q50,q75 = np.quantile(kratio, [0.25,.5,0.75])
print(q25,q50,q75, sep='\t')
iqr = q75 - q25
floor = q25-2*iqr
print(floor)
nonemptyk = np.where(kratio > floor)[0]

data = raw.iloc[nonemptyk, :]
print(data.shape)

filename = src + 'traditional_phenotype_summary.csv'
print(filename)
data.to_csv(filename, index=False)

data.head()
```

```python
ucaccsd = sorted(np.unique(UCACCSD))[1:]
print(len(ucaccsd))

ucaccsd = sorted(np.unique(data['UCACCSD']))[1:]
print(len(ucaccsd))
```

```python
quant = np.zeros((len(ucaccsd), data.shape[1]-3 + attr.shape[1] - 5 + 2))
i = 0
for i in range(len(quant)):
    uc = ucaccsd[i]
    
    foo = data[data['UCACCSD'] == uc]
    bar = attr[attr['UCACCSD'] == uc]
    
    quant[i] = np.hstack(([len(foo), len(bar)], 
                          np.nanmean(foo.iloc[:,3:].values, axis = 0),
                          np.nanmean(bar.iloc[:,5:].values, axis = 0)))
```

```python
col = ['nut_count', 'plot_count'] + list(data.columns[3:]) + list(attr.columns[5:])
```

```python
qual = pd.DataFrame(quant, columns = col)
qual = qual.astype({'nut_count': int, 'plot_count':int})
foo = np.asarray([qual.shape[1]] + list(range(qual.shape[1])))

qual['UCACCSD'] = ucaccsd

qual = qual.iloc[:, foo]
print(qual.shape)

filename = src + 'qual_quant_summary.csv'
print(filename)
qual.to_csv(filename, index=False)

qual.head()
```
