```python
src = '../hpcc/traditional/'

col_names = ['bname', 'wname',
             'nut_length','nut_height', 'nut_width',
             'nut_vol',
             'nut_va3d',
             'nut_ratio_feret',
             'nut_area',
             'nut_sphericity',
             'nut_cnvx_area',
             'nut_cnvx_vol',
             'nut_ratio_cnvx_area',
             'nut_inv_ratio_cnvx_vol',
             'krumbein',
             'sneed',
             'air_vol', 'meat_vol', 'shell_vol', 'vein_vol',
             'air_ratio_vol', 'meat_ratio_vol', 'shell_ratio_vol', 'vein_ratio_vol',
             'shell_rugosity',
             'shell_thickness',
             'shell_ext_ratio', 'shell_prot_ratio',
             'shell_ext_vol', 'shell_prot_vol',
             'meat_length', 'meat_height', 'meat_width',
             'meat_area',
             'meat_cnvx_vol',
             'meat_cnvx_area',
             'meat_ratio_cnvx_area',
             'meat_inv_ratio_cnvx_vol',
             'meat_cavity_area',
             'meat_cavity_vol',
             'meat_arc_height',
             'meat_cavity_height',
             'meat_sphericity',
             'meat_ratio_cavity_area',
             'meat_ratio_cavity_vol',
             'rho_mvs', 'rho_vvs', 'rho_vvm'
             ]

filename = src + 'traditional_raw.csv'
data = pd.read_csv(filename, header=None, dtype={1:str}, names=col_names)
data.loc[data['bname'] == '201SBa_R5_T55', 'bname'] = '2014SBa_R5_T55'
foo = data.iloc[:, 30:33].values
MM = np.max(foo, axis=1)
mm = np.min(foo, axis=1)
data['meat_ratio_feret'] = MM/mm
#data = data.drop(columns = [''])
print(data.shape)
data.head()
```

```python
for i in range(len(data.columns)):
    print(i,data.columns[i], sep='\t')
```

```python
cattraits = np.zeros((len(data), 15)) - 1
UCACCSD = []

for i in range(len(cattraits)):
    locate, Rrow, Ttree = data.iloc[i,0].split('_')
    row = int(Rrow[1:])
    tree = int(Ttree[1:])
    
    if locate == 'NewStuke':
        locate = 'New Stuke'
    #elif locate in ['SelectC','SelectD','SelectF']:
    elif locate in ['SelectD', 'SelectF', 'SelectC']:
        locate = 'Select Davis'

    metadf = attr[(attr['LOCATE']==locate) & (attr['ROW']==row) & (attr['TREE']==tree)]
    if len(metadf) > 0:
        meta = metadf.iloc[0]
        cattraits[i] = meta.iloc[5:].values.astype(float)
        UCACCSD.append(metadf.iloc[0,0])
    else:
        #print(data.iloc[i,0])
        UCACCSD.append('')
        pass

foo = pd.DataFrame(cattraits, columns=attr.columns[5:])
uc = pd.DataFrame(UCACCSD, columns=['UCACCSD'])
df = pd.concat([uc, data, foo], axis=1)
print(df.shape)
df.head()
```

```python
WID = []
NN = []
for wid in np.sort(pd.unique(df['UCACCSD']))[1:]:
    foo = df[df.UCACCSD == wid]
    WID.append(wid)
    NN.append(len(foo))
foo = pd.DataFrame()
foo['UCACCSD'] = WID
foo['N'] = np.asarray(NN, dtype=int)
foo.to_csv('../scanned_accessions.csv', index=False)
```

```python
j = 21
kratio = data.iloc[:,j].values
q25,q50,q75 = np.quantile(kratio, [0.25,.5,0.75])
print(q25,q50,q75, sep='\t')
iqr = q75 - q25
floor = q25-2*iqr
print(floor)
nonemptyk = np.where(kratio > floor)[0]

df = df.iloc[nonemptyk, :]
print(df.shape)

filename = src + 'traditional_phenotype_summary.csv'
print(filename)
```

```python
df.to_csv(filename, index=False)
```
